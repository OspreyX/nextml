```julia
# Problem 1. Parallel Monte Carlo Longest Increasing Subsequence

# Use a parallel monte carlos simulation to generate a histogram of the longest increasing subsequences of some big random
# permutations (see Julia function `randperm`).
# The resulting distribution is understood, but not widely known.

# Note that the value of a `@parallel` loop body can be anything, and the reduction
# function can also be any (associative) function.
```

```julia
@everywhere function patiencesort(p::Vector)
    piles = similar(p, 0)
    for pi in p
        idx = 1
        for pp in piles
            if pi > pp
                idx += 1
            end
        end
        if idx > length(piles)
            d = length(piles)
            resize!(piles, idx)
            piles[d+1:end] = 0
        end
        piles[idx] = pi
    end
    return length(piles)
end
```
